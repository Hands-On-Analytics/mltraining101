```python
import matplotlib.pyplot as plt

plt.hist(np.random.rand(100000),bins=100,histtype="step",density=True,color="red",label="rand") #uniform distribution
plt.hist(np.random.randn(100000),bins=100,histtype="step",density=True,color="green",label="randn") #normal distribution
plt.legend(loc="upper left")
plt.title("Random distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axis([-2.5,2.5,0,1.1])
plt.show()
```
